```python
df = pd.read_csv('/content/test_scores.csv')
df
```

```python
def GD(x,y):
  mc_curr = bc_curr = 0
  iteration = 1000
  n = len(x)
  learning_rate = 0.0001
  for i in range(iteration):
    y_predicted = mc_curr*x + bc_curr
    md = (2/n)*sum(-x*(y-y_predicted))
    bd = -(2/n)*sum(y-y_predicted)
    mc_curr = mc_curr  - learning_rate*md
    bc_curr = bc_curr - learning_rate*bd
    cost = (1/n)*sum([val**2 for val in y- y_predicted ])

    print("m {}, b{}, iteration {}, cost{}".format(mc_curr, bc_curr,cost, i))
    
```

```python
x = df.math
y = df.cs
GD(x,y)
```
